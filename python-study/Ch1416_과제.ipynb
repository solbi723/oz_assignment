[문제3]

숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"  
234567 → "23four5six7"  
10203 → "1zerotwozero3"    

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,   
혹은 바뀌지 않고 그대로인 문자열 s가 주어졌을때
s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

---
참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어  
0	zero  
1	one  
2	two  
3	three  
4	four  
5	five  
6	six  
7	seven  
8	eight  
9	nine  

```python
def convert_to_number(s):
    
    word_to_digit = {
        "zero": "0",
        "one": "1",
        "two": "2",
        "three": "3",
        "four": "4",
        "five": "5",
        "six": "6",
        "seven": "7",
        "eight": "8",
        "nine": "9"
    }
    
    
    result = ""
    buffer = ""  

    for char in s:
        if char.isdigit():  
            if buffer: 
                result += word_to_digit[buffer]  
                buffer = ""  
            result += char  
        else: 
            buffer += char  


    if buffer:
        result += word_to_digit[buffer]

    return result


s = "77three4one"

original_number = convert_to_number(s)
print(original_number)
```

[문제4]

행렬의 덧셈은 행과 열의 크기가 같은 두 행렬의 같은 행, 같은 열의 값을 서로 더한 결과가 됩니다. 2개의 행렬 arr1과 arr2가 주어졌을 때, 행렬 덧셈의 결과를 출력하는 코드를 작성해주세요(result와 같은 결과가 출력되도록 작성)

arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = [[4,6],[7,9]]

```python
def matrix_addition(arr1, arr2):
    
    rows = len(arr1)
    cols = len(arr1[0])
    
    result = [[0] * cols for _ in range(rows)]
    

    for i in range(rows):
        for j in range(cols):
            result[i][j] = arr1[i][j] + arr2[i][j]
    
    return result

arr1 = [[1, 2], [2, 3]]
arr2 = [[3, 4], [5, 6]]

result = matrix_addition(arr1, arr2)

print(result)
```

[문제7]

0부터 9까지의 숫자 중 일부가 들어있는 정수 배열 numbers가 주어집니다.   
numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 출력해주세요

numbers = [5,8,1,0,6,9]

```python
def sum_missing_numbers(numbers):

    all_numbers = set(range(10))
 
    numbers_set = set(numbers)

    missing_numbers = all_numbers - numbers_set

    return sum(missing_numbers)

numbers = [5, 8, 1, 0, 6, 9]

result = sum_missing_numbers(numbers)
result
```

[문제 1: 붕어빵 재고 확인]

붕어빵 가게의 현재 재고를 확인하는 기능을 구현하세요.

요구사항:

- 딕셔너리에는 붕어빵의 종류와 각각의 재고 개수가 저장됩니다.
- 현재 재고를 화면에 출력하세요.

붕어빵 재고
- 팥붕어빵 : 10개
- 슈크림붕어빵 : 8개
- 초코붕어빵 : 5개

```python
bungeobbang_stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

def check_stock(stock):
    print("현재 붕어빵 재고:")
    for kind, quantity in stock.items():
        print(f"{kind} : {quantity}개")

check_stock(bungeobbang_stock)
```

[문제 2: 붕어빵 주문 받기]

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
def bungeobbang_order():
    order = {}  # 주문 내용을 저장할 딕셔너리

    while True:
        kind = input("주문할 붕어빵의 종류를 입력하세요 (종료하려면 '끝' 입력): ")
        if kind == '끝':
            break
        qty = input(f"{kind}의 개수를 입력하세요: ")
        
        try:
            qty = int(qty)
            if qty < 0:
                print("개수는 0 이상이어야 합니다.")
                continue
        except ValueError:
            print("유효한 개수를 입력하세요.")
            continue
        
        order[kind] = order.get(kind, 0) + qty

    print("\n주문 내역:")
    for kind, qty in order.items():
        print(f"{kind}: {qty}개")

# 함수 호출
bungeobbang_order()
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
bungeobbang_stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

def check_stock(stock):
    order = input("주문할 붕어빵 종류와 개수를 입력하세요 (예: 팥붕어빵 2): ")
    kind, qty = order.split()
    qty = int(qty)

  
    if kind in stock:
       
        if stock[kind] >= qty:
            
            stock[kind] -= qty
            print(f"{kind} {qty}개 판매 완료!")
        else:
            print(f"{kind}의 재고가 부족합니다. 현재 재고: {stock[kind]}개")
    else:
        print(f"{kind}는 재고에 없습니다.")


    print("현재 붕어빵 재고:")
    for kind, qty in stock.items():
        print(f"{kind} : {qty}개")


check_stock(bungeobbang_stock)
```

[문제 5: 관리자 모드]

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
def admin_mode(stock):
    while True:
        command = input("추가할 붕어빵 종류와 개수를 입력하세요 (예: 팥붕어빵 5) 또는 '종료'를 입력하세요: ")
        if command == "종료":
            print("관리자 모드를 종료합니다.")
            break
        try:
            kind, qty = command.split()
            qty = int(qty)
        except ValueError:
            print("잘못된 입력입니다. 다시 시도하세요.")
            continue
        
        if kind in stock:
            stock[kind] += qty
            print(f"{kind}의 재고가 {qty}개 추가되었습니다. 현재 재고: {stock[kind]}개")
        else:
            print(f"{kind}는 재고에 없습니다. 새로운 붕어빵을 추가하려면 재고 목록에 추가하세요.")

        print("현재 붕어빵 재고:")
        for kind, qty in stock.items():
            print(f"{kind} : {qty}개")

while True:
    mode = input("주문 모드 또는 관리자 모드로 들어가시겠습니까? (주문/관리자): ")
    if mode == "주문":
        check_stock(bungeobbang_stock)
    elif mode == "관리자":
        admin_mode(bungeobbang_stock)
    else:
        print("잘못된 입력입니다. '주문' 또는 '관리자'를 입력하세요.")

```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
bungeobbang_stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

bungeobbang_prices = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}

sales_count = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

def check_stock(stock, prices, sales):
    total_revenue = 0  # 총 매출 초기화
    while True:
        order = input("주문할 붕어빵 종류와 개수를 입력하세요 (예: 팥붕어빵 2) 또는 '종료'를 입력하세요: ")
        
        if order == "종료":
            print("주문을 종료합니다.")
            break
        
        try:
            kind, qty = order.split()
            qty = int(qty)
        except ValueError:
            print("잘못된 입력입니다. 다시 시도하세요.")
            continue
        
        if kind in stock:
            if stock[kind] >= qty:
                stock[kind] -= qty
                sales[kind] += qty  # 판매 개수 증가
                revenue = prices[kind] * qty  # 매출 계산
                total_revenue += revenue  # 총 매출에 더하기
                print(f"{kind} {qty}개 판매 완료! 매출: {revenue}원")
            else:
                print(f"{kind}의 재고가 부족합니다. 현재 재고: {stock[kind]}개")
        else:
            print(f"{kind}는 재고에 없습니다.")
        
        print("현재 붕어빵 재고:")
        for kind, qty in stock.items():
            print(f"{kind} : {qty}개")
        
        print("현재 판매 개수:")
        for kind, qty in sales.items():
            print(f"{kind} : {qty}개")

    print(f"총 매출: {total_revenue}원")  # 총 매출 출력

def admin_mode(stock):
    while True:
        command = input("추가할 붕어빵 종류와 개수를 입력하세요 (예: 팥붕어빵 5) 또는 '종료'를 입력하세요: ")
        if command == "종료":
            print("관리자 모드를 종료합니다.")
            break
        try:
            kind, qty = command.split()
            qty = int(qty)
        except ValueError:
            print("잘못된 입력입니다. 다시 시도하세요.")
            continue
        
        if kind in stock:
            stock[kind] += qty
            print(f"{kind}의 재고가 {qty}개 추가되었습니다. 현재 재고: {stock[kind]}개")
        else:
            print(f"{kind}는 재고에 없습니다. 새로운 붕어빵을 추가하려면 재고 목록에 추가하세요.")

        print("현재 붕어빵 재고:")
        for kind, qty in stock.items():
            print(f"{kind} : {qty}개")

while True:
    mode = input("주문 모드 또는 관리자 모드로 들어가시겠습니까? (주문/관리자): ")
    if mode == "주문":
        check_stock(bungeobbang_stock, bungeobbang_prices, sales_count)
    elif mode == "관리자":
        admin_mode(bungeobbang_stock)
    else:
        print("잘못된 입력입니다. '주문' 또는 '관리자'를 입력하세요.")
```
