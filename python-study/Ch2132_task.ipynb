사용자로부터 이름과 나이를 입력 받아, "안녕하세요, [이름]님! 당신은 [나이]살입니다."라고 출력하는 함수 greet_user를 작성하세요.

```python
def greet_user():
    name = input("이름을 입력하세요: ")
    age = input("나이를 입력하세요: ")
    print(f"안녕하세요, {name}님! 당신은 {age}살입니다.")

# 함수 실행 예시
greet_user()
```

여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.


```python
def sum_numbers(*numbers):
    return sum(numbers)

result = sum_numbers(1, 2, 3, 4, 5)
print(result)  
```

0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n: int):
    for i in range(n + 1):
        if i % 2 != 0:
            yield i
```

사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max():

    nums = tuple(int(input(f"숫자 {i+1}을 입력하세요: ")) for i in range(3))

    print(f"입력된 튜플: {nums}")
    print(f"최소값: {min(nums)}")
    print(f"최대값: {max(nums)}")

min_max()
```

주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(num: int) -> bool:
    if num % 2 == 0:
        return True
    return False
```

팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n: int) -> int:
    if n == 0 or n == 1:  
        return 1
    return n * factorial(n - 1)  

print(factorial(6))  
```

문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(strings: list[str]) -> None:
    for string in strings:
        print(f"{string}: {len(string)}")

print_lengths(["마라탕", "곱창", "요아정먹고싶다"])  
```

두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(str1: str, str2: str) -> str:
    if len(str1) >= len(str2):
        return str1
    return str2

# 사용 예시
print(longer_string("파인애플", "사과"))  
print(longer_string("고양이", "강아지"))   
print(longer_string("박솔비", "바"))       
```
